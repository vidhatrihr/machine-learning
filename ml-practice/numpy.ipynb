# axis in numpy



```python
x = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]


array = np.array(x)

axis = array.sum(axis=1)
axis

array.ndim

array.nbytes  # total bytes consumed


array.argmax()  # index where the elem is maximum

array.argmin()  # index where the elem is minimum

array.argsort()  # gives index of each elem
```

```python
lst = [3, 4, 0, 2, 1, 30, 5]
maxi = lst[0]
mini = lst[0]
for x in lst:
  if x > maxi:
    maxi = x
  if x < mini:
    mini = x
print(maxi, mini)
```
