```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.array([
    [1, 0.5],
    [2, 2.3],
    [3, 3.1],
    [4, 3.9],
])


def encode1(x):  # x is a vector
  return x[0] - x[1]


def decode1(u):   # u is scalar
  return np.array([u/2, u/2])


loss = 0
for x in data:
  u = encode1(x)
  x_decoded = decode1(u)
  loss += sum((x - x_decoded) ** 2)

loss / 4
```

```python
def encode2(x):
  return (x[0]+x[1])/2


def decode2(u):
  return np.array([u/2, u/2])


loss = 0
for x in data:
  u = encode2(x)
  x_decoded = decode2(u)
  loss += sum((x-x_decoded) ** 2)

loss / 4
```

```python
data = np.array([
    [-1, 0.0319],
    [0, 0.8692],
    [1, 1.9566],
    [2, 3.0343],
])

x = data[:, 0]
y = data[:, 1]

params = [[1, 1], [1, 2], [2, 1], [2, 2]]


mini_mse = float('inf')
best_param = 0, 0
for a, b in params:
  y_pred = x * a + b
  mse = sum((y-y_pred)**2) / len(x)
  if mse < mini_mse:
    mini_mse = mse
    best_param = a, b

print(best_param, mini_mse)
```

```python
data = np.array([
    [1, 2, 3],
    [2, 3, 4],
    [-1, 0, 1],
    [0, 1, 1]
])


def encoder(x):
  return (x[0]+2*x[1]) / 2


def decoder(u):
  return np.array([u, 2*u, 3*u])


loss = 0
for x in data:
  u = encoder(x)
  x_decoded = decoder(u)
  loss += sum((x - x_decoded) ** 2)
  x_decoded

loss / len(data)
```
