```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
array = np.array([10, 20, 30, 40, 40])
n = len(array)
```

```python
mean = sum(array)/n
mean
```

```python
sorted_arr = sorted(array)

if n % 2 == 1:
  median = sorted_arr[n//2]
else:
  median = (sorted_arr[n//2 - 1] + sorted_arr[n//2]) / 2

median
```

```python
freq = {}

for num in array:
  if num not in freq:
    freq[num] = 1
  else:
    freq[num] += 1
```

```python
mode_num = array[0]
mode_freq = 1

for num in array:
  if freq[num] > mode_freq:
    mode_freq = freq[num]
    mode_num = num

mode_num
```

```python
number = 1.345/2.4579
print(f'{number:.2f}')
```

```python
def plot_stats():
  plt.figure(figsize=(6, 4))

  # plt.scatter(range(len(array)), array, color='m', label='Values', marker='s')

  plt.plot(array, marker='X', label='Values')
  plt.axhline(mean, color='r', label=f'{mean=:.2f}', linestyle='-.')
  plt.axhline(median, color='g', label=f'{median=:.2f}', linestyle='--')
  plt.axhline(mode_num, color='b', label=f'{mode_num=:.2f}', linestyle=':')

  plt.xlabel('Indices')
  plt.ylabel('Values')
  plt.title('Statistics')
  plt.grid(True)
  plt.legend()


plot_stats()
```
